```python
### Import statements
# For generating dummy training data
from pyDOE2 import *
import numpy as np
# For gaussian process regressors
from sklearn.gaussian_process.kernels import Matern
from sklearn.gaussian_process import GaussianProcessRegressor
from math import sin, cos, pi
# For plotting
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Winkler Method for Fusion of Information Sources
- This notebook presents the Winkler method for the fusion of multiple information sources, represented as Gaussian Processes.
- The method follows the procedure in Winkler RL (1981) Combining probability distributions from depen- dent information sources. Manag Sci 27(4):479â€“488
- Two Gaussian Processes will be created and fused using the Winkler Method. (Just like in the actual optimization problem)
- It is assumed that the design space is 3-dimensional and the objective space is 1-dimensional. (Just like in the actual optimization problem)

```python
### Define Projection Function 
# This function takes the samples out of the LHS function and transforms them into uniform samples in the actual design space
def sample_project(x, a_range, b_range, c_range):
    n_samp = len(x)
    
    a_lower = a_range[0]
    a_upper = a_range[1]
    
    b_lower = b_range[0]
    b_upper = b_range[1]
    
    c_lower = c_range[0]
    c_upper = c_range[1]
    
    a_samples_projected = x[:,0]*(a_upper - a_lower) + np.ones(n_samp)*a_lower
    b_samples_projected = x[:,1]*(b_upper - b_lower) + np.ones(n_samp)*b_lower
    c_samples_projected = x[:,2]*(c_upper - c_lower) + np.ones(n_samp)*c_lower
    
    x_projected = np.column_stack((a_samples_projected, b_samples_projected, c_samples_projected))
    
    return x_projected
```

```python
### Generate Dummy Training data
# Assumed bounds on the 3 dimensions x = [a,b,c]
a_low = -10
a_up = 10

b_low = -5
b_up = 5

c_low = 0
c_up = 50

# Define Number of training samples
n_samples = 200

# Generate Training Samples using Latin Hypercube Sampling
x_lhs1 = lhs(3, samples=n_samples, criterion='centermaximin') # For GP1
x_lhs2 = lhs(3, samples=n_samples, criterion='centermaximin') # For GP2

# Project the samples into the actual design space
x_proj1 = sample_project(x_lhs1, [a_low, a_up], [b_low, b_up], [c_low, c_up]) # For GP1
x_proj2 = sample_project(x_lhs2, [a_low, a_up], [b_low, b_up], [c_low, c_up]) # For GP2

#print('x_lhs1')
#print(x_lhs1)
#print('x_proj1')
#print(x_proj1)

# Plotting generated samples
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')
ax1.scatter(x_proj1[:,0], x_proj1[:,1], x_proj1[:,2], marker='o')
ax1.set_xlabel('Design paramter a in units')
ax1.set_ylabel('Design paramter b in units')
ax1.set_zlabel('Design paramter c in units')
ax1.title.set_text('Samples for GP1 using Latin Hypercube Sampling')
plt.show()

fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(x_proj2[:,0], x_proj2[:,1], x_proj2[:,2], marker='o')
ax2.set_xlabel('Design paramter a in units')
ax2.set_ylabel('Design paramter b in units')
ax2.set_zlabel('Design paramter c in units')
ax2.title.set_text('Samples for GP2 using Latin Hypercube Sampling')
plt.show()
```

## Create and train Gaussian Processes
- Now that the training samples for the two GPs have been generated. the two GPs are trained
- The first GP is trained to simulate function f_1(x) = -exp(b)(a - 3)sin(c*pi)
- The second GP is trained to simulated function f_2(x) = -exp((10-b)/2)(a - 5)cos(c*pi)
- The Matern Kernel is used for both GPs
- The hyperparameters are estimated using Maximization of the Log Marginal Likelihood using the L-BFGS algorithm

```python
### Define the two functions
def func_1(x):
    a = x[0]
    b = x[1]
    c = x[2]
    
    y_1 = -np.exp(b)*(a - 3)*sin(c*pi)
    return y_1

def func_2(x):
    a = x[0]
    b = x[1]
    c = x[2]
    
    y_2 = -np.exp((10-b)/2)*(a - 5)*cos(c*pi)
    return y_2
```

```python
### Visualizing the two functions
c_vec = np.linspace(c_low, c_up, 3)
a_vec = np.linspace(a_low, a_up, n_samples)
b_vec = np.linspace(b_low, b_up, n_samples)
a_vals, b_vals = np.meshgrid(a_vec, b_vec)
    
fig3 = plt.figure()
for i in range(len(c_vec)):
    y_GP1 = np.zeros([len(a_vec),len(b_vec)])
    for j in range(len(a_vec)):
        for k in range(len(b_vec)):
            y_GP1[j][k] = func_1([a_vals[j][k], b_vals[j][k], c_vec[i]])
    # Plot Function f_1(x)
    ax3 = fig3.add_subplot(1,len(c_vec),i+1, projection='3d')
    ax3.plot_surface(a_vals, b_vals, y_GP1, cmap='magma')
    ax3.set_title(['f_1(x) for c = ' + str(c_vec[i]) + ' units'])
    ax3.set_xlabel('a in units')
    ax3.set_ylabel('b in units')
    ax3.set_zlabel('f_1(x) in units')
plt.show()
   
fig4 = plt.figure()
for i in range(len(c_vec)):
    y_GP2 = np.zeros([len(a_vec),len(b_vec)])
    for j in range(len(a_vec)):
        for k in range(len(b_vec)):
            y_GP2[j][k] = func_2([a_vals[j][k], b_vals[j][k], c_vec[i]])
    # Plot Function f_2(x)
    ax4 = fig4.add_subplot(1,len(c_vec),i+1, projection='3d')
    ax4.plot_surface(a_vals, b_vals, y_GP2, cmap='magma')
    ax4.set_title(['f_2(x) for c = ' + str(c_vec[i]) + ' units'])
    ax4.set_xlabel('a in units')
    ax4.set_ylabel('b in units')
    ax4.set_zlabel('f_2(x) in units')
plt.show()
```

```python
### Find the noisy outputs for the two sample datasets and train the GPs
y_GP1 = np.zeros(n_samples)
y_GP2 = np.zeros(n_samples)

y_GP1_train = np.zeros(n_samples)
y_GP2_train = np.zeros(n_samples)

for i in range(n_samples):
    x_curr_1 = x_proj1[i,:]
    x_curr_2 = x_proj2[i,:]
    
    y_GP1[i] = func_1(x_curr_1)
    y_GP2[i] = func_2(x_curr_2)
    
    y_GP1_train[i] = y_GP1[i] + 0.1*np.random.random_sample()
    y_GP2_train[i] = y_GP2[i] + 0.1*np.random.random_sample() 
    
len_scale = [a_up*2/n_samples, b_up*2/n_samples, c_up*2/n_samples]
kernel = Matern(length_scale=len_scale, nu=2.5)
GP1_prior = GaussianProcessRegressor(kernel=kernel, random_state=1)
GP2_prior = GaussianProcessRegressor(kernel=kernel, random_state=2)

GP1_post = GaussianProcessRegressor(kernel=kernel, random_state=1).fit(x_proj1, y_GP1_train)
GP2_post = GaussianProcessRegressor(kernel=kernel, random_state=2).fit(x_proj2, y_GP2_train)

## Test predictive capability of GPs
y_GP1_prior_mean = np.zeros(n_samples)
#y_GP1_prior_std = np.zeros(n_samples)
y_GP1_post_mean = np.zeros(n_samples)
#y_GP1_post_std = np.zeros(n_samples)

y_GP2_prior_mean = np.zeros(n_samples)
#y_GP2_prior_std = np.zeros(n_samples)
y_GP2_post_mean = np.zeros(n_samples)
#y_GP2_post_std = np.zeros(n_samples)

y_GP1_prior_mean  = GP1_prior.predict(x_proj1)
y_GP1_post_mean  = GP1_post.predict(x_proj1)

#y_GP1_prior_mean, y_GP1_prior_std  = GP1_prior.predict(x_proj1, return_std=True)
#y_GP1_post_mean, y_GP1_post_std  = GP1_post.predict(x_proj1, return_std=True)

y_GP2_prior_mean = GP2_prior.predict(x_proj2)
y_GP2_post_mean = GP2_post.predict(x_proj2)

#y_GP2_prior_mean, y_GP2_prior_std = GP2_prior.predict(x_proj2, return_std=True)
#y_GP2_post_mean, y_GP2_post_std = GP2_post.predict(x_proj2, return_std=True)

## Compute model fidelity based variance
var_fid1 = np.mean(np.subtract(y_GP1, y_GP1_post_mean))**2
var_fid2 = np.mean(np.subtract(y_GP2, y_GP2_post_mean))**2

fig5 = plt.figure()
ax1 = fig5.add_subplot(1,2,1)
ax1.plot(y_GP1, y_GP1_prior_mean, 'bo')
ax1.plot(y_GP1, y_GP1, 'k-')
ax1.set_xlabel('True Function Values in units')
ax1.set_ylabel('Predictions from prior GP in units')
ax2 = fig5.add_subplot(1,2,2)
ax2.plot(y_GP1, y_GP1_post_mean, 'bo')
ax2.plot(y_GP1, y_GP1, 'k-')
ax2.set_xlabel('True Function Values in units')
ax2.set_ylabel('Predictions from posterior GP in units')
plt.show()

fig6 = plt.figure()
ax1 = fig6.add_subplot(1,2,1)
ax1.plot(y_GP2, y_GP2_prior_mean, 'bo')
ax1.plot(y_GP2, y_GP2, 'k-')
ax1.set_xlabel('True Function Values in units')
ax1.set_ylabel('Predictions from prior GP in units')
ax2 = fig6.add_subplot(1,2,2)
ax2.plot(y_GP2, y_GP2_post_mean, 'bo')
ax2.plot(y_GP2, y_GP2, 'k-')
ax2.set_xlabel('True Function Values in units')
ax2.set_ylabel('Predictions from posterior GP in units')
plt.show()
```

## Perform Winkler Fusion
- Once the GPs are trained, Winkler Fusion is performed
- Winkler Fusion uses a different set of design samples. This design sample set is also generated using Latin Hypercube Sampling
- The Winkler Means for the samples are taken as the reference response values to train the fused GP

```python
### Generate sample dataset for fusion process
# Define Number of samples for fusion
n_samples_fusion = 100

# Generate Fusion Samples using Latin Hypercube Sampling
x_fuse_lhs = lhs(3, samples=n_samples_fusion, criterion='centermaximin') 

# Project the samples into the actual design space
x_fuse_proj = sample_project(x_fuse_lhs, [a_low, a_up], [b_low, b_up], [c_low, c_up])

### Define the function that computes the Winkler means
def Winkler_mean(x, gp_1, gp_2, sigma_f1, sigma_f2):
    n_des = len(x)
    
    gp1_means = np.zeros(n_des)
    gp1_stds = np.zeros(n_des)
    
    gp2_means = np.zeros(n_des)
    gp2_stds = np.zeros(n_des)
    
    gp_means = np.zeros([n_des,2])
    gp_vars = np.zeros([n_des,2])
    
    #gp_1 = gp_array[0]
    #gp_2 = gp_array[1]
    
    [gp1_means, gp1_stds] = gp_1.predict(x, return_std=True)
    [gp2_means, gp2_stds] = gp_2.predict(x, return_std=True)
    
    gp_means = np.column_stack((gp1_means, gp2_means))
    
    var_1 = gp1_stds**2 + sigma_f1
    var_2 = gp2_stds**2 + sigma_f2
    
    gp_vars = np.column_stack((var_1, var_2))
    
    rho_tilde = np.zeros([n_des,2,2])
    rho = np.zeros([n_des,2,2])
    covar = np.zeros([n_des,2,2])
    mean_winkler = np.zeros(n_des)
    #print(rho_tilde[0].shape)
    
    ## Compute rho_tilde matrix for current design sample
    for i in range(n_des):
        #rho_tilde_x = np.zeros([2,2])
        mean_vec = gp_means[i,:]
        var_vec = gp_vars[i,:]
        for j1 in range(2):
            for k1 in range(2):
                mean_j = mean_vec[j1]
                mean_k = mean_vec[k1]
                var_j = var_vec[j1]
                var_k = var_vec[k1]
                #rho_tilde_x[j1,k1] = np.sqrt(var_j)/(np.sqrt((mean_j - mean_k)**2 + var_j))
                rho_tilde[i,j1,k1] = np.sqrt(var_j)/(np.sqrt((mean_j - mean_k)**2 + var_j))
    
    ## Compute correlation matrix for current design sample
    for i2 in range(n_des):
        rho_tilde_x = rho_tilde[i2,:,:]
        #rho_x = np.zeros([2,2])
        var_vec = gp_vars[i2,:]
        for j2 in range(2):
            for k2 in range(2):
                var_j = var_vec[j2]
                var_k = var_vec[k2]
                #rho_x[j2,k2] = (var_k/(var_j + var_k))*rho_tilde_x[j2,k2] + (var_j/(var_j + var_k))*rho_tilde_x[k2,j2]
                rho[i2,j2,k2] = (var_k/(var_j + var_k))*rho_tilde_x[j2,k2] + (var_j/(var_j + var_k))*rho_tilde_x[k2,j2]
        
    ## Compute covariance matrix for current design sample
    for i3 in range(n_des):
        rho_x = rho[i3,:,:]
        #covar_x = np.zeros([2,2])                                             
        var_vec = gp_vars[i3,:]                                             
        for j3 in range(2):
            for k3 in range(2):
                std_j = np.sqrt(var_vec[j3])
                std_k = np.sqrt(var_vec[k3])
                #covar_x[j3,k3] = rho_x[j3,k3]*std_j*std_k
                covar[i3,j3,k3] = rho_x[j3,k3]*std_j*std_k
        
    ## Compute Winkler Mean
    e_vec = np.ones(2)
    for i4 in range(n_des):
        mean_vec = gp_means[i4,:]
        covar_x = covar[i4,:,:]
        mean_winkler[i4] = (np.linalg.multi_dot([np.transpose(e_vec), np.linalg.inv(covar_x), mean_vec]))/(np.linalg.multi_dot([np.transpose(e_vec), np.linalg.inv(covar_x), e_vec]))
                             
    return [mean_winkler, covar]

## Compute the Winkler means for the generated design samples
[y_wink, covar_models] = Winkler_mean(x_fuse_proj, GP1_post, GP2_post, var_fid1, var_fid2)
```

```python
### Train the fused GP using the Winkler means
len_scale = [a_up*2/n_samples, b_up*2/n_samples, c_up*2/n_samples]
kernel_fused = Matern(length_scale=len_scale, nu=2.5)
GP_fused = GaussianProcessRegressor(kernel=kernel_fused, random_state=1)

GP_fused_post = GP_fused.fit(x_fuse_proj, y_wink)

## Compute function values for comparison
y_func1 = np.zeros(n_samples_fusion)
y_func2 = np.zeros(n_samples_fusion)
for i in range(n_samples_fusion):
    x_curr = x_fuse_proj[i,:]
    y_func1[i] = func_1(x_curr_1)
    y_func2[i] = func_2(x_curr_2)

### Plot Fused GP predictions with function values
fig7 = plt.figure()
ax1 = fig7.add_subplot(1,2,1)
ax1.plot(y_func1, y_wink, 'bo')
ax1.plot(y_func1, y_func1, 'k-')
ax1.set_xlabel('Function 1 Values in units')
ax1.set_ylabel('Predictions from fused GP in units')
ax2 = fig7.add_subplot(1,2,2)
ax2.plot(y_func2, y_wink, 'bo')
ax2.plot(y_func2, y_func2, 'k-')
ax2.set_xlabel('Function 2 Values in units')
ax2.set_ylabel('Predictions from fused GP in units')
plt.show()
```
