# Regressor Comparison using Analytical Model
- Three Regressors: Neural Network, Gaussian Process and Linear Regressor will be compared
- Design samples will be generated using Latin Hypercube Sampling 
- The generated designs will be evaluated using the Analytical Buckling Model

```python
### Analytical Model
def Pcr_Analytical(x,l,w,t,E_f,n_eigen):
    r_f = x[0] # in mm
    theta_f = x[1] # in degrees
    E_m = x[2] # in GPa
    #m = 1
    #n = 1
    P_cr = []
    poisson_ratio = 0.3
    v_f = volume_fraction(l,w,t,x)
    ## Rule of Mixtures
    E_effective = (E_f*v_f + (1-v_f)*E_m)*1e3 # Converting to N/mm^2
    D = E_effective*t**3/(12*(1-poisson_ratio**2)) # Flexural Rigidity
    for i in range(n_eigen):
        m = i+1
        n = 1
        P_cr.append(D*(pi*l/m)**2*((m/l)**2+(n/w)**2)**2)
    return P_cr
    
### Volume Fraction
def volume_fraction(length,width,thickness,x_design):
    r_f = x_design[0]
    theta_f = x_design[1]
    v_total = length*width*thickness
    l_fibre_shortest = length/(6*cos(radians(theta_f)))
    l_fibre_med = 2*l_fibre_shortest
    l_fibre_longest = 3*l_fibre_shortest
    vol_fibres = pi*(r_f**2)*(2*l_fibre_shortest + 2*l_fibre_med + l_fibre_longest)
    vol_total = length*width*thickness
    vol_frac = vol_fibres/vol_total
    return vol_frac
```

## Generate Training Dataset
- Latin Hypercube Sampling is used to generate uniformly distributed samples.
- Each generated dataset is checked for satisfaction of the volume fraction constraint.
- NOTE: There is no random seed argument for Latin Hypercube Sampling Function "lhs". Therefore rerunning the cell will provide a different set of design samples. An email was sent to the author to clarify this, but no response was received. 

```python
### Define the number of designs for Neural Network training (DO NOT Run Again!)
n_train = 100

### Define maximum limit for volume fraction
vf_limit = 0.5

### Define the Problem Constants
l_matrix = 3048 # in mm
w_matrix = l_matrix # in mm
t_matrix = 254 # in mm
E_fibre = 350 # in GPa

### Define the input design parameter limits
Em_upper_limit = 800 # in GPa
Em_lower_limit = 10 # in GPa

rf_upper_limit = t_matrix/2 # in mm
rf_lower_limit = 25 # in mm

theta_f_upper_limit = 90 # in deg
theta_f_lower_limit = -90 # in deg

### Forming Gaussian Process Surrogate
def project(x, Em_lower_limit, Em_upper_limit, rf_lower_limit, rf_upper_limit, theta_f_lower_limit, theta_f_upper_limit,):
    # x = [rf, theta_f, Em] * n_train
    rf_projected = x[:,0]*(rf_upper_limit-rf_lower_limit) + rf_lower_limit
    theta_f_projected = x[:,1]*(theta_f_upper_limit-theta_f_lower_limit) + theta_f_lower_limit
    Em_projected = x[:,2]*(Em_upper_limit-Em_lower_limit) + Em_lower_limit
    x_projected = np.vstack((rf_projected, theta_f_projected, Em_projected)).T
    return x_projected

conv = False
n_iter = 0
while(not conv):
    x_gen = lhs(3, samples=n_train, criterion='maximin')
    x_gen_proj = project(x_gen, Em_lower_limit, Em_upper_limit, rf_lower_limit, rf_upper_limit, theta_f_lower_limit, theta_f_upper_limit)    
    for i in range(n_train):
        v_f = volume_fraction(l_matrix, w_matrix, t_matrix, x_gen_proj[i,:])
        if (v_f>vf_limit):
            break
        elif (i==n_train-1):
            conv = True
    n_iter = n_iter + 1
    
#print(x_train_proj)
print(n_iter)

### Plotting the sampled datapoints
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x_gen_proj[:,0], x_gen_proj[:,1], x_gen_proj[:,2],marker='o')
ax.set_xlabel('Fibre Radius in mm')
ax.set_ylabel('Fibre Angle in degrees')
ax.set_zlabel('Matrix E in GPa')
fig.savefig('lhs_sampling_modelcomp.png')
```

## Train-Test Split
- The generated designs are evaluated using the Analytical Model
- The data is then pre-processed, i.e. centred and scaled to unit variance 
- The preprocessed data is then split into a training dataset and testing dataset

```python
### Evaluating generated designs with Analytical Model
n_eigen = 1 # Number of requested Buckling Eigen Values
y_analytical = np.zeros([n_train,n_eigen])
for i in range(n_train):
    y_analytical[i,:] = Pcr_Analytical(x_gen_proj[i,:],l_matrix,w_matrix,t_matrix,E_fibre,n_eigen)
    
### Data preprocessing
scaler = StandardScaler()
scaler.fit(y_analytical.reshape(-1,1))
print(scaler.mean_)
y_scaled = scaler.transform(y_analytical.reshape(-1,1))    

### Splitting into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(x_gen_proj, y_scaled, test_size=0.2, random_state=104) 
```

## Neural Network Training
- Using the training dataset, the Neural Network is trained
- A cross-validation dataset is taken from the training data
- The Keras library is used for the Neural Network

```python
## Defining the Neural Network
Pcr_NN = Sequential()
Pcr_NN.add(Dense(15, activation='relu', input_dim=3))
Pcr_NN.add(Dropout(0.3))
Pcr_NN.add(Dense(30, activation='relu'))
Pcr_NN.add(Dropout(0.3))
Pcr_NN.add(Dense(10, activation='relu'))
Pcr_NN.add(Dropout(0.3))
Pcr_NN.add(Dense(1))

#sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
Pcr_NN.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])

## Training the Neural Network using the training dataset
Pcr_History = Pcr_NN.fit(X_train, y_train, batch_size=32, epochs=100, validation_split=0.2)
    
## Plotting the Mean Squared Error of the training and validation datasets
fig2 = plt.figure()
plt.plot(Pcr_History.history['mse'])
plt.plot(Pcr_History.history['val_mse'])
plt.title('Neural Net Mean Squared Error')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.legend(['Training','Validation'],loc='upper right')
plt.show()
fig2.savefig('Neural_Net_train.png')

## Show and save a figure of the Neural Network
#SVG(model_to_dot(Pcr_NN, show_shapes=True).create(prog='dot', format='svg')) # To display in the notebook, however image is too big
plot_model(Pcr_NN, show_shapes=True, to_file='Neural_Net.png') # same image as SVG
```

## Gaussian Process Regressor Training
- The same training dataset is used to train a Gaussian Process Regressor
- The Matern Kernel is used for covariance computation
- To show that the GPR is fit well, a small subset of the training dataset is used to compare the GPR prediction with the actual value
- The printed GPR score is the coefficient of determination R^2 metric. The best possible value is 1.0
- NOTE: since the GPR predicts the buckling load for the trained designs very well, only the predicted mean and true values are plotted

```python
# Training the GPR
custom_kernel = Matern()
GPR = GaussianProcessRegressor(kernel=custom_kernel,random_state=10).fit(X_train, y_train)

## Validation
n_val = 25 # number of validation points
X_val = X_train[:n_val,:]
y_val = y_train[:n_val]
[y_gpr_mean, y_gpr_std] = GPR.predict(X_val, return_std=True)
print(GPR.score(X_val, y_val))

x_gpr = np.arange(1,n_val+1)
fig2 = plt.figure()
plt.plot(x_gpr, y_gpr_mean, 'b-')
#plt.plot(x_gpr, y_gpr_mean+3*y_gpr_std, 'r-')
#plt.plot(x_gpr, y_gpr_mean-3*y_gpr_std, 'k-')
plt.plot(x_gpr, y_val, 'g-')
plt.xlabel('Validation datapoint')
plt.ylabel('Buckling Load')
plt.title('GPR Prediction')
#plt.legend(['GPR mean','GPR std upper','GPR std lower','truth'],loc='upper right')
plt.legend(['GPR mean','truth'],loc='upper right')
plt.show()
fig2.savefig('GPR_pred.png')
```

## Linear Regressor Training
- The same training dataset is used to train the linear regressor
- Similar to the GPR, Validation will be performed using a subset of the training dataset 

```python
## Training the linear Regressor
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

## Printing the slope and intercept of trained linear model
print(linear_model.intercept_)
print(linear_model.coef_)

## Validation
n_val_lin = 25
X_val_lin = X_train[:n_val_lin,:]
y_val_lin = y_train[:n_val_lin]
y_lin_pred = linear_model.predict(X_val_lin)

x_lin = np.arange(1,n_val_lin+1)
fig3 = plt.figure()
plt.plot(x_lin, y_lin_pred, 'b-')
plt.plot(x_lin, y_val_lin, 'g-')
plt.xlabel('Validation datapoint')
plt.ylabel('Buckling Load')
plt.title('Linear Regressor Prediction')
plt.legend(['LR prediction','truth'],loc='upper right')
plt.show()
fig3.savefig('LinMod_pred.png')
```

## Testing the trained models with testing dataset
- The three trained models will now be tested on the testing dataset
- The means and standard deviations of the models will be plotted as a box plot
- Since the Neural Network and Linear Regressor do not naturally provide an uncertainty value, the error computed from the true values will be used

```python
## Determining predictions from each model for the testing dataset
y_nn_test = Pcr_NN.predict(X_test) # Neural Network
[y_gpr_test_mean, y_gpr_test_std] = GPR.predict(X_test, return_std=True) # GPR
y_lin_test = linear_model.predict(X_test) # Linear Regressor

## Compute errors for Neural Net and Linear Regressor
y_nn_err = y_nn_test - y_test
y_lin_err = y_lin_test - y_test

## Comparison Box Plot
plot_data = [y_nn_test.reshape(-1,), y_gpr_test_std, y_lin_err.reshape(-1,)]
#data_array = np.array(plot_data)
#print(np.shape(data_array))
names = ['Neural Network', 'GPR', 'Linear Regressor']
fig4 = plt.figure()
ax = fig4.add_subplot(111)
plt.boxplot(plot_data, showfliers=False)
plt.ylabel('Uncertainty metric')
ax.set_xticklabels(names)
plt.show()
fig4.savefig('Model_comparison.png')
```
