```python
### Import statements
# For generating dummy training data
from pyDOE2 import *
import numpy as np
# For gaussian process regressors
from sklearn.gaussian_process.kernels import Matern
from sklearn.gaussian_process import GaussianProcessRegressor
from math import sin
# For plotting
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Winkler Method for Fusion of Information Sources
- This notebook presents the Winkler method for the fusion of multiple information sources, represented as Gaussian Processes.
- The method follows the procedure in Winkler RL (1981) Combining probability distributions from depen- dent information sources. Manag Sci 27(4):479â€“488
- Two Gaussian Processes will be created and fused using the Winkler Method. (Just like in the actual optimization problem)
- The design space and objective space are considered to be one dimensional.

```python
### Define Projection Function 
# This function takes the samples out of the LHS function and transforms them into uniform samples in the actual design space
def sample_project(x, x_range):
    n_samp = len(x)
    
    x_lower = x_range[0]
    x_upper = x_range[1]
    
    x_projected = np.zeros(n_samp)
    for i in range(n_samp):
        x_projected[i] = x[i]*(x_upper - x_lower) + x_lower
    
    return x_projected
```

```python
### Generate Dummy Training data
# Assumed bounds on x
x_low = -10
x_up = 10

# Define Number of training samples
n_samples = 50

# Generate Training Samples using Latin Hypercube Sampling
x_lhs1 = lhs(1, samples=n_samples, criterion='centermaximin') # For GP1
x_lhs2 = lhs(1, samples=n_samples, criterion='centermaximin') # For GP2

# Project the samples into the actual design space
x_proj1 = sample_project(x_lhs1, [x_low, x_up]) # For GP1
x_proj2 = sample_project(x_lhs2, [x_low, x_up]) # For GP2

x_vec = np.linspace(1,n_samples,n_samples)
# Plotting generated samples
fig1 = plt.figure()
plt.scatter(x_vec, x_proj1, marker='o')
plt.xlabel('Design point')
plt.ylabel('Design paramter value in units')
plt.title('Samples for GP1 using Latin Hypercube Sampling')
plt.show()

fig2 = plt.figure()
plt.scatter(x_vec, x_proj2, marker='o')
plt.xlabel('Design point')
plt.ylabel('Design paramter value in units')
plt.title('Samples for GP2 using Latin Hypercube Sampling')
plt.show()
```

## Create and train Gaussian Processes
- Now that the training samples for the two GPs have been generated. the two GPs are trained
- The first GP is trained to simulate function f_1(x) = -(1.6 - 3*x)*sin(18*x)
- The second GP is trained to simulated function f_2(x) = -(1.8 - 3*x)*sin(18*x + 0.1)
- The Matern Kernel is used for both GPs
- The hyperparameters are estimated using Maximization of the Log Marginal Likelihood using the L-BFGS algorithm

```python
### Define the two functions
def func_1(x):
    y_1 = -(1.6 - 3*x)*(sin(18*x))
    return y_1

def func_2(x):
    y_2 = -(1.8 - 3*x)*sin(18*x + 0.1)
    return y_2
```

```python
### Visualizing the two functions
x = np.linspace(0,1.2,50)
y1 = np.zeros(len(x))
y2 = np.zeros(len(x))
for i in range(len(x)):
    y1[i] = func_1(x[i])
    y2[i] = func_2(x[i])

fig3 = plt.figure()
plt.plot(x,y1)
plt.xlabel('x-value in units')
plt.ylabel('y=f_1(x) in units')
plt.show()

fig4 = plt.figure()
plt.plot(x,y2)
plt.xlabel('x-value in units')
plt.ylabel('y=f21(x) in units')
plt.show()
```

```python
### Find the noisy outputs for the two sample datasets and train the GPs
y_f1 = np.zeros(n_samples)
y_f2 = np.zeros(n_samples)

y_gp1_train = np.zeros(n_samples)
y_gp2_train = np.zeros(n_samples)

for i in range(n_samples):
    x_curr_1 = x_proj1[i]
    x_curr_2 = x_proj2[i]
    
    y_f1[i] = func_1(x_curr_1)
    y_f2[i] = func_2(x_curr_2)
    
    y_gp1_train[i] = y_f1[i] + 0.1*np.random.random_sample()
    y_gp2_train[i] = y_f2[i] + 0.1*np.random.random_sample() 
    
len_scale = [x_up*2/n_samples]
kernel = Matern(length_scale=len_scale, nu=2.5)
GP1_prior = GaussianProcessRegressor(kernel=kernel, random_state=1)
GP2_prior = GaussianProcessRegressor(kernel=kernel, random_state=2)

GP1_post = GaussianProcessRegressor(kernel=kernel, random_state=1).fit(x_proj1.reshape(-1,1), y_gp1_train)
GP2_post = GaussianProcessRegressor(kernel=kernel, random_state=2).fit(x_proj2.reshape(-1,1), y_gp2_train)

## Test predictive capability of GPs
y_GP1_prior_mean = np.zeros(n_samples)
y_GP1_post_mean = np.zeros(n_samples)

y_GP2_prior_mean = np.zeros(n_samples)
y_GP2_post_mean = np.zeros(n_samples)

y_GP1_prior_mean  = GP1_prior.predict(x_proj1.reshape(-1,1))
y_GP1_post_mean  = GP1_post.predict(x_proj1.reshape(-1,1))

y_GP2_prior_mean = GP2_prior.predict(x_proj2.reshape(-1,1))
y_GP2_post_mean = GP2_post.predict(x_proj2.reshape(-1,1))

## Compute model fidelity based variance
var_fid1 = np.mean(np.subtract(y_f1, y_GP1_post_mean))**2
var_fid2 = np.mean(np.subtract(y_f2, y_GP2_post_mean))**2

fig5 = plt.figure()
ax1 = fig5.add_subplot(1,2,1)
ax1.plot(y_f1, y_GP1_prior_mean, 'bo')
ax1.plot(y_f1, y_f1, 'k-')
ax1.set_xlabel('True Function Values in units')
ax1.set_ylabel('Predictions from prior GP in units')
ax2 = fig5.add_subplot(1,2,2)
ax2.plot(y_f1, y_GP1_post_mean, 'bo')
ax2.plot(y_f1, y_f1, 'k-')
ax2.set_xlabel('True Function Values in units')
ax2.set_ylabel('Predictions from posterior GP in units')
plt.show()

fig6 = plt.figure()
ax1 = fig6.add_subplot(1,2,1)
ax1.plot(y_f2, y_GP2_prior_mean, 'bo')
ax1.plot(y_f2, y_f2, 'k-')
ax1.set_xlabel('True Function Values in units')
ax1.set_ylabel('Predictions from prior GP in units')
ax2 = fig6.add_subplot(1,2,2)
ax2.plot(y_f2, y_GP2_post_mean, 'bo')
ax2.plot(y_f2, y_f2, 'k-')
ax2.set_xlabel('True Function Values in units')
ax2.set_ylabel('Predictions from posterior GP in units')
plt.show()
```

- It can be seen that the prior is characterized by zero mean
- The posterior, once its trained, perfectly predicts the responses for the designs it has been trained over

## Perform Winkler Fusion
- Once the GPs are trained, Winkler Fusion is performed
- Winkler Fusion uses a different set of design samples. This design sample set is also generated using Latin Hypercube Sampling
- The Winkler Means for the samples are taken as the reference response values to train the fused GP

```python
### Generate sample dataset for fusion process
# Define Number of samples for fusion
n_samples_fusion = 60

# Generate Fusion Samples using Latin Hypercube Sampling
x_fuse_lhs = lhs(1, samples=n_samples_fusion, criterion='centermaximin') 

# Project the samples into the actual design space
x_fuse_proj = sample_project(x_fuse_lhs, [x_low, x_up])

### Define the function that computes the Winkler means
def Winkler_fusion(x, gp_1, gp_2, sigma_f1, sigma_f2):
    n_des = len(x)
    
    gp1_means = np.zeros(n_des)
    gp1_stds = np.zeros(n_des)
    gp2_means = np.zeros(n_des)
    gp2_stds = np.zeros(n_des)
    gp_means = np.zeros([n_des,2])
    gp_vars = np.zeros([n_des,2])
    
    [gp1_means, gp1_stds] = gp_1.predict(x.reshape(-1,1), return_std=True)
    [gp2_means, gp2_stds] = gp_2.predict(x.reshape(-1,1), return_std=True)
    gp_means = np.column_stack((gp1_means, gp2_means))
    
    var_1 = gp1_stds**2 + sigma_f1
    var_2 = gp2_stds**2 + sigma_f2
    gp_vars = np.column_stack((var_1, var_2))
    
    rho_tilde = np.zeros([n_des,2,2])
    rho = np.zeros([n_des,2,2])
    covar = np.zeros([n_des,2,2])
    mean_winkler = np.zeros(n_des)
    var_winkler = np.zeros(n_des)
    
    for i in range(n_des):
        # Compute rho_tilde matrix for current design sample
        rho_tilde_x = np.zeros([2,2])
        mean_vec = gp_means[i,:]
        var_vec = gp_vars[i,:]
        for j1 in range(2):
            for k1 in range(2):
                mean_j = mean_vec[j1]
                mean_k = mean_vec[k1]
                var_j = var_vec[j1]
                var_k = var_vec[k1]
                rho_tilde_x[j1,k1] = np.sqrt(var_j)/(np.sqrt((mean_j - mean_k)**2 + var_j))
    
        # Compute correlation matrix for current design sample
        rho_x = np.zeros([2,2])
        #var_vec = gp_vars[i,:]
        for j2 in range(2):
            for k2 in range(2):
                var_j = var_vec[j2]
                var_k = var_vec[k2]
                rho_x[j2,k2] = (var_k/(var_j + var_k))*rho_tilde_x[j2,k2] + (var_j/(var_j + var_k))*rho_tilde_x[k2,j2]
        
        # Compute covariance matrix for current design sample
        covar_x = np.zeros([2,2])                                             
        for j3 in range(2):
            for k3 in range(2):
                std_j = np.sqrt(var_vec[j3])
                std_k = np.sqrt(var_vec[k3])
                covar_x[j3,k3] = rho_x[j3,k3]*std_j*std_k
        
        # Compute Winkler Mean
        e_vec = np.ones(2)
        mean_winkler[i] = (np.linalg.multi_dot([np.transpose(e_vec), np.linalg.inv(covar_x), mean_vec]))/(np.linalg.multi_dot([np.transpose(e_vec), np.linalg.inv(covar_x), e_vec]))
        var_winkler[i] = 1/(np.linalg.multi_dot([np.transpose(e_vec), np.linalg.inv(covar_x), e_vec]))
        rho_tilde[i,:,:] = rho_tilde_x
        rho[i,:,:] = rho_x
        covar[i,:,:] = covar_x
    
    return mean_winkler, var_winkler

## Compute the Winkler means for the generated design samples
[y_wink, var_wink] = Winkler_fusion(x_fuse_proj, GP1_post, GP2_post, var_fid1, var_fid2)
```

```python
### Train the fused GP using the Winkler means
len_scale = [x_up*2/n_samples]
kernel_fused = Matern(length_scale=len_scale, nu=2.5)
GP_fused = GaussianProcessRegressor(kernel=kernel_fused, random_state=1)

GP_fused_post = GP_fused.fit(x_fuse_proj.reshape(-1,1), y_wink)

## Compute function values for comparison
y_func1 = np.zeros(n_samples_fusion)
y_func2 = np.zeros(n_samples_fusion)
for i in range(n_samples_fusion):
    x_curr = x_fuse_proj[i]
    y_func1[i] = func_1(x_curr)
    y_func2[i] = func_2(x_curr)

### Plot Fused GP predictions with function values
fig7 = plt.figure()
ax1 = fig7.add_subplot(1,2,1)
ax1.plot(y_func1, y_wink, 'bo')
ax1.plot(y_func1, y_func1, 'k-')
ax1.set_xlabel('Function 1 Values in units')
ax1.set_ylabel('Predictions from fused GP in units')
ax2 = fig7.add_subplot(1,2,2)
ax2.plot(y_func2, y_wink, 'bo')
ax2.plot(y_func2, y_func2, 'k-')
ax2.set_xlabel('Function 2 Values in units')
ax2.set_ylabel('Predictions from fused GP in units')
plt.show()
```

- It can be seen that the prediction of the fused GP is scattered around the true function values.
- However, the fused mean prediction will be closer to either of the function values depending upon the uncertainty in the predictions of both GPs. This is captured in the correlation matrix
- Also, it is worth mentioning that the more training datapoints available for both the GPs, the better the fused prediction wiil be.
